```python
import os
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Path to converted WAV files
wav_path = "raw_audio/converted_wav"

# Load one sample file
sample_file = os.listdir(wav_path)[0]
audio_path = os.path.join(wav_path, sample_file)

y, sr = librosa.load(audio_path, sr=16000)

# Feature extraction
duration = len(y) / sr
rms = np.mean(librosa.feature.rms(y=y))
zcr = np.mean(librosa.feature.zero_crossing_rate(y))
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
mfcc_mean = np.mean(mfcc, axis=1)

# Display results
print("Loaded File:", sample_file)
print("Sample Rate:", sr)
print("Duration (seconds):", round(duration, 2))
print("RMS Energy:", round(rms, 5))
print("Zero Crossing Rate:", round(zcr, 5))
print("\nMFCC Features:")
print(mfcc_mean)

# Plot waveform
plt.figure(figsize=(10, 3))
plt.plot(y)
plt.title("Respiratory Audio Waveform")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.show()

# Prepare feature dictionary (app-ready)
features = {
    "duration": duration,
    "rms": rms,
    "zcr": zcr
}

for i, val in enumerate(mfcc_mean):
    features[f"mfcc_{i+1}"] = val

print("\nFeature Dictionary (for App Integration):")
print(features)
```

```python
import streamlit as st
import librosa
import numpy as np
import matplotlib.pyplot as plt
import tempfile
import os

# ---------- PAGE CONFIG ----------
st.set_page_config(
    page_title="BreathCare",
    page_icon="🫁",
    layout="centered"
)

# ---------- HEADER ----------
st.markdown("<h1 style='text-align: center;'>🫁 BreathCare</h1>", unsafe_allow_html=True)
st.markdown("<p style='text-align: center;'>Respiratory Audio Analysis System</p>", unsafe_allow_html=True)
st.markdown("---")

# ---------- FILE UPLOAD ----------
st.subheader("Upload Respiratory Audio")
uploaded_file = st.file_uploader("Upload WAV audio file", type=["wav"])

if uploaded_file is not None:
    # Save temp file
    with tempfile.NamedTemporaryFile(delete=False) as tmp:
        tmp.write(uploaded_file.read())
        audio_path = tmp.name

    # Audio Player
    st.audio(uploaded_file, format="audio/wav")

    # Load audio
    y, sr = librosa.load(audio_path, sr=16000)

    # ---------- FEATURE EXTRACTION ----------
    duration = round(len(y) / sr, 2)
    rms = round(np.mean(librosa.feature.rms(y=y)), 5)
    zcr = round(np.mean(librosa.feature.zero_crossing_rate(y)), 5)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    mfcc_mean = np.mean(mfcc, axis=1)

    # ---------- DISPLAY FEATURES ----------
    st.markdown("### Extracted Audio Features")

    col1, col2, col3 = st.columns(3)
    col1.metric("Duration (sec)", duration)
    col2.metric("RMS Energy", rms)
    col3.metric("ZCR", zcr)

    # ---------- MFCC TABLE ----------
    st.markdown("### MFCC Values")
    mfcc_data = {f"MFCC {i+1}": round(val, 3) for i, val in enumerate(mfcc_mean)}
    st.json(mfcc_data)

    # ---------- WAVEFORM ----------
    st.markdown("### Audio Waveform")
    fig, ax = plt.subplots(figsize=(8, 3))
    ax.plot(y)
    ax.set_xlabel("Samples")
    ax.set_ylabel("Amplitude")
    ax.set_title("Respiratory Audio Waveform")
    st.pyplot(fig)

    # ---------- FOOTER ----------
    st.markdown("---")
    st.success("Audio analysis completed successfully")

else:
    st.info("Please upload a WAV audio file to begin analysis.")
```
