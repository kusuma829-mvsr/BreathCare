```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Load dataset
df = pd.read_csv("final_breathcare_dataset.csv")

# Drop non-numeric / unwanted columns
X = df.drop(columns=["filename"], errors="ignore")

# Handle missing values
X = X.dropna()

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("Dataset cleaned & scaled successfully")
print("Final shape:", X_scaled.shape)
```

```python
print("Mean (first 5 features):")
print(X_scaled.mean().head())

print("\nStandard Deviation (first 5 features):")
print(X_scaled.std().head())
```

```python
X_scaled.to_csv("breathcare_scaled_dataset.csv", index=False)
print("Final scaled dataset saved")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.imshow(X_scaled.corr())
plt.colorbar()
plt.title("Feature Correlation Heatmap")
plt.show()
```
