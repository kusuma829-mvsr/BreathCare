```python
import os
from pydub import AudioSegment

input_path = "raw_audio/wav_data"
output_path = "raw_audio/converted_wav"
os.makedirs(output_path, exist_ok=True)

audio_files = [f for f in os.listdir(input_path) if f.endswith((".webm", ".ogg"))]

print("Audio files found:", len(audio_files))

success = 0
fail = 0

for file in audio_files:
    try:
        audio = AudioSegment.from_file(os.path.join(input_path, file))
        wav_name = file.rsplit(".", 1)[0] + ".wav"
        audio.export(os.path.join(output_path, wav_name), format="wav")
        success += 1
    except:
        fail += 1

print("Conversion finished")
print("Successful:", success)
print("Failed:", fail)
```
