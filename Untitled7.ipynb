```python
import pandas as pd

df = pd.read_csv("combined_dataset.csv")
print(df.columns)
```

```python
import numpy as np

# Example: assign random labels (for testing only)
df["label"] = np.random.choice(
    ["normal", "asthma", "pneumonia", "bronchitis"],
    size=len(df)
)

print(df["label"].value_counts())
```

```python
df.to_csv("final_labeled_dataset.csv", index=False)
```

```python
print(df["label"].value_counts())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop(columns=["filename", "label"])
y = df["label"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=200, random_state=42)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
import joblib

joblib.dump(model, "breath_model.pkl")
joblib.dump(scaler, "scaler.pkl")
```

```python
import librosa
import numpy as np

def extract_features(file_path):
    y, sr = librosa.load(file_path, sr=16000)

    duration = len(y) / sr
    rms = np.mean(librosa.feature.rms(y=y))
    zcr = np.mean(librosa.feature.zero_crossing_rate(y))

    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    mfcc_mean = np.mean(mfcc, axis=1)

    features = [duration, rms, zcr] + mfcc_mean.tolist()
    return np.array(features).reshape(1, -1)
```
