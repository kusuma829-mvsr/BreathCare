```python
import librosa
import numpy as np
import os

wav_path = "raw_audio/converted_wav"
sample_file = os.listdir(wav_path)[0]

y, sr = librosa.load(os.path.join(wav_path, sample_file), sr=16000)

print("Loaded:", sample_file)
print("Duration (sec):", round(len(y)/sr, 2))
```

```python
def extract_features(file_path):
    y, sr = librosa.load(file_path, sr=16000)

    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    mfcc_mean = np.mean(mfcc, axis=1)

    rms = np.mean(librosa.feature.rms(y=y))
    zcr = np.mean(librosa.feature.zero_crossing_rate(y))
    duration = len(y) / sr

    return mfcc_mean, rms, zcr, duration
```

```python
import pandas as pd

data = []

for file in os.listdir(wav_path):
    if file.endswith(".wav"):
        try:
            mfcc, rms, zcr, duration = extract_features(
                os.path.join(wav_path, file)
            )

            row = {
                "filename": file,
                "duration": duration,
                "rms": rms,
                "zcr": zcr
            }

            for i, val in enumerate(mfcc):
                row[f"mfcc_{i+1}"] = val

            data.append(row)

        except:
            continue

df = pd.DataFrame(data)
df.to_csv("combined_dataset.csv", index=False)

print("CSV created successfully!")
print("Total samples:", len(df))
```
