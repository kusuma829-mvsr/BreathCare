```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Load sample audio (no error version)
y, sr = librosa.load(librosa.ex('trumpet'), sr=16000)

# -------- AUDIO PREPROCESSING --------
# Normalize audio
y = librosa.util.normalize(y)

# Plot waveform
plt.figure(figsize=(10,4))
plt.plot(y)
plt.title("Preprocessed Audio Waveform")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()


# -------- FEATURE EXTRACTION --------

# MFCC
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

# RMS Energy
rms = librosa.feature.rms(y=y)

# Zero Crossing Rate
zcr = librosa.feature.zero_crossing_rate(y)

# Plot MFCC
plt.figure(figsize=(10,4))
plt.imshow(mfcc, aspect='auto', origin='lower')
plt.title("MFCC Features")
plt.colorbar()
plt.xlabel("Time")
plt.ylabel("MFCC Coefficients")
plt.show()

# Print values
print("RMS Value:", np.mean(rms))
print("ZCR Value:", np.mean(zcr))
```
